```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Vector Addition Visualization
def plot_vector_addition():
    """Visualize vector addition in R2"""
    fig, ax = plt.subplots(figsize=(10, 6))

    # Define vectors
    v1 = np.array([2, 1])
    v2 = np.array([1, 2])
    v_sum = v1 + v2

    # Plot vectors
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1 = (2,1)')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2 = (1,2)')
    ax.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='v1 + v2 = (3,3)')

    # Plot the parallelogram
    ax.plot([0, v1[0], v_sum[0], v2[0], 0], [0, v1[1], v_sum[1], v2[1], 0], 'k--')

    # Set limits and labels
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 4)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Vector Addition: The Parallelogram Law')
    ax.grid(True)
    ax.legend()

    plt.show()

# 2. Scalar Multiplication Visualization
def plot_scalar_multiplication():
    """Visualize scalar multiplication in R2"""
    fig, ax = plt.subplots(figsize=(10, 6))

    # Define vector
    v = np.array([2, 1])

    # Define scalars
    scalars = [-1, 0.5, 1, 2]
    colors = ['r', 'g', 'b', 'purple']

    # Plot vectors
    for scalar, color in zip(scalars, colors):
        scaled_v = scalar * v
        ax.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1,
                  color=color, label=f'{scalar}v')

    # Set limits and labels
    ax.set_xlim(-3, 5)
    ax.set_ylim(-3, 3)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Scalar Multiplication')
    ax.grid(True)
    ax.legend()

    plt.show()

# 3. Linear Independence vs Dependence Visualization
def plot_linear_independence():
    """Visualize linear independence vs dependence in R3"""
    fig = plt.figure(figsize=(15, 12))

    # First subplot: linearly independent vectors
    ax1 = fig.add_subplot(121, projection='3d')
    v1 = np.array([1, 0, 0])
    v2 = np.array([0, 1, 0])
    v3 = np.array([0, 0, 1])

    # Plot vectors
    ax1.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1')
    ax1.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='g', label='v2')
    ax1.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='b', label='v3')

    ax1.set_xlim(-1, 1.5)
    ax1.set_ylim(-1, 1.5)
    ax1.set_zlim(-1, 1.5)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('z')
    ax1.set_title('Linearly Independent Vectors')

    # Second subplot: linearly dependent vectors
    ax2 = fig.add_subplot(122, projection='3d')
    v1 = np.array([1, 0, 0])
    v2 = np.array([0, 1, 0])
    v3 = np.array([1, 1, 0])  # v3 = v1 + v2 (dependent)

    # Plot vectors
    ax2.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1')
    ax2.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='g', label='v2')
    ax2.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='b', label='v3 = v1 + v2')

    ax2.set_xlim(-1, 1.5)
    ax2.set_ylim(-1, 1.5)
    ax2.set_zlim(-1, 1.5)
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('z')
    ax2.set_title('Linearly Dependent Vectors')

    plt.tight_layout()
    plt.show()

# 4. Spanning Set Visualization
def plot_spanning_set():
    """Visualize a spanning set in R2"""
    fig, ax = plt.subplots(figsize=(8, 8))

    # Define basis vectors
    v1 = np.array([1, 0])
    v2 = np.array([0, 1])

    # Plot grid of linear combinations
    for i in range(-3, 4):
        for j in range(-3, 4):
            point = i*v1 + j*v2
            ax.scatter(point[0], point[1], color='blue', s=15)

    # Plot basis vectors
    ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
    ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')

    # Set limits and labels
    ax.set_xlim(-3.5, 3.5)
    ax.set_ylim(-3.5, 3.5)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Spanning Set: Points Generated by Linear Combinations')
    ax.grid(True)
    ax.legend()

    plt.show()

# 5. Subspace Visualization
def plot_subspace():
    """Visualize subspaces in R3"""
    fig = plt.figure(figsize=(15, 5))

    # First subplot: line subspace
    ax1 = fig.add_subplot(131, projection='3d')
    # Line: span{(1,1,1)}
    t = np.linspace(-2, 2, 100)
    x = t
    y = t
    z = t

    ax1.plot(x, y, z, 'b', label='Line Subspace')
    ax1.set_xlim(-2, 2)
    ax1.set_ylim(-2, 2)
    ax1.set_zlim(-2, 2)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('z')
    ax1.set_title('Line Through Origin')

    # Second subplot: plane subspace
    ax2 = fig.add_subplot(132, projection='3d')
    # Plane: span{(1,0,0), (0,1,0)}
    xx, yy = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
    zz = np.zeros_like(xx)

    ax2.plot_surface(xx, yy, zz, alpha=0.5, color='g')
    ax2.set_xlim(-2, 2)
    ax2.set_ylim(-2, 2)
    ax2.set_zlim(-2, 2)
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('z')
    ax2.set_title('Plane Subspace')

    # Third subplot: all of R3
    ax3 = fig.add_subplot(133, projection='3d')
    # Show axes
    ax3.quiver(0, 0, 0, 2, 0, 0, color='r', label='x-axis')
    ax3.quiver(0, 0, 0, 0, 2, 0, color='g', label='y-axis')
    ax3.quiver(0, 0, 0, 0, 0, 2, color='b', label='z-axis')

    ax3.set_xlim(-2, 2)
    ax3.set_ylim(-2, 2)
    ax3.set_zlim(-2, 2)
    ax3.set_xlabel('x')
    ax3.set_ylabel('y')
    ax3.set_zlabel('z')
    ax3.set_title('Full Space R³')

    plt.tight_layout()
    plt.show()

# Call the functions to display the visualizations
# Uncomment these lines to run the visualizations
plot_vector_addition()
plot_scalar_multiplication()
plot_linear_independence()
plot_spanning_set()
plot_subspace()
```
