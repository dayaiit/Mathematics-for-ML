```python
# Example 1: Basic Linear Mappings
# This example demonstrates how to verify if a mapping is linear or not

import numpy as np
import matplotlib.pyplot as plt

```

```python
# ---- PART 1: DEFINING LINEAR MAPPINGS ----

# A linear mapping F: V → W must satisfy two properties:
# 1. F(v + w) = F(v) + F(w)   [Preserves addition]
# 2. F(k*v) = k*F(v)         [Preserves scalar multiplication]

# Let's define some simple mappings and check if they're linear

def mapping_A(v):
    """Linear mapping: F(x, y) = (2x + y, x - y)"""
    return np.array([2*v[0] + v[1], v[0] - v[1]])

def mapping_B(v):
    """Non-linear mapping: G(x, y) = (x^2, y)"""
    return np.array([v[0]**2, v[1]])

def mapping_C(v):
    """Non-linear mapping: H(x, y) = (x + 1, y)"""
    return np.array([v[0] + 1, v[1]])
```

```python

# ---- PART 2: VERIFYING LINEARITY ----

# Let's verify the linearity of mapping_A
print("Verifying linearity of mapping A: F(x, y) = (2x + y, x - y)")

# Choose two test vectors
v1 = np.array([2, 3])
v2 = np.array([1, 4])
print(f"Vector v1 = {v1}")
print(f"Vector v2 = {v2}")

# Test property 1: F(v1 + v2) = F(v1) + F(v2)
sum_then_map = mapping_A(v1 + v2)
map_then_sum = mapping_A(v1) + mapping_A(v2)

print("\nProperty 1: F(v1 + v2) = F(v1) + F(v2)")
print(f"F(v1 + v2) = F({v1 + v2}) = {sum_then_map}")
print(f"F(v1) + F(v2) = {mapping_A(v1)} + {mapping_A(v2)} = {map_then_sum}")
print(f"Are they equal? {np.array_equal(sum_then_map, map_then_sum)}")

# Test property 2: F(k*v) = k*F(v)
k = 3  # Choose a scalar
scaled_then_map = mapping_A(k * v1)
map_then_scaled = k * mapping_A(v1)

print("\nProperty 2: F(k*v) = k*F(v)")
print(f"F(k*v1) = F({k}*{v1}) = F({k*v1}) = {scaled_then_map}")
print(f"k*F(v1) = {k}*{mapping_A(v1)} = {map_then_scaled}")
print(f"Are they equal? {np.array_equal(scaled_then_map, map_then_scaled)}")

print("\nSince both properties hold, mapping A is linear.")
print()
```

```python

# ---- PART 3: CHECKING NON-LINEAR MAPPINGS ----

# Let's verify that mapping_B is not linear
print("Verifying that mapping B: G(x, y) = (x^2, y) is not linear")

# Test property 1: G(v1 + v2) = G(v1) + G(v2)
sum_then_map = mapping_B(v1 + v2)
map_then_sum = mapping_B(v1) + mapping_B(v2)

print("\nProperty 1: G(v1 + v2) = G(v1) + G(v2)")
print(f"G(v1 + v2) = G({v1 + v2}) = {sum_then_map}")
print(f"G(v1) + G(v2) = {mapping_B(v1)} + {mapping_B(v2)} = {map_then_sum}")
print(f"Are they equal? {np.array_equal(sum_then_map, map_then_sum)}")
print("Since Property 1 doesn't hold, mapping B is not linear.")
print()

# Let's verify that mapping_C is not linear
print("Verifying that mapping C: H(x, y) = (x + 1, y) is not linear")

# Check if H(0) = 0 (a quick test for linearity)
zero_vector = np.array([0, 0])
print(f"H(0, 0) = {mapping_C(zero_vector)}")
print("Since H(0, 0) ≠ (0, 0), mapping C is not linear.")
print()
```

```python

# ---- PART 4: VISUALIZATION OF LINEAR MAPPINGS ----

# Let's visualize what the linear mapping A does to a grid of points
plt.figure(figsize=(15, 6))

# Plot 1: Original grid
plt.subplot(1, 2, 1)
x = np.linspace(-2, 2, 5)
y = np.linspace(-2, 2, 5)
X, Y = np.meshgrid(x, y)

plt.plot(X, Y, 'b-')
plt.plot(X.T, Y.T, 'b-')
plt.scatter(X, Y, c='r')

# Mark a few specific vectors
plt.scatter([v1[0]], [v1[1]], c='green', s=100, label='v1')
plt.scatter([v2[0]], [v2[1]], c='purple', s=100, label='v2')
plt.scatter([v1[0]+v2[0]], [v1[1]+v2[1]], c='orange', s=100, label='v1+v2')

plt.grid(True)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Original Grid')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Plot 2: Transformed grid
plt.subplot(1, 2, 2)

# Transform each point and plot
for i in range(len(X)):
    for j in range(len(X[0])):
        point = np.array([X[i,j], Y[i,j]])
        transformed = mapping_A(point)
        X[i,j] = transformed[0]
        Y[i,j] = transformed[1]

plt.plot(X, Y, 'b-')
plt.plot(X.T, Y.T, 'b-')
plt.scatter(X, Y, c='r')

# Transform and mark the specific vectors
v1_transformed = mapping_A(v1)
v2_transformed = mapping_A(v2)
v1_plus_v2_transformed = mapping_A(v1 + v2)

plt.scatter([v1_transformed[0]], [v1_transformed[1]], c='green', s=100, label='F(v1)')
plt.scatter([v2_transformed[0]], [v2_transformed[1]], c='purple', s=100, label='F(v2)')
plt.scatter([v1_plus_v2_transformed[0]], [v1_plus_v2_transformed[1]], c='orange', s=100, label='F(v1+v2)')

plt.grid(True)
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.title('Transformed Grid (Linear Mapping A)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()

# ---- EXERCISES FOR STUDENTS ----

# 1. Define and verify linearity for the mapping F(x, y, z) = (x + y, y - z, x + z)

# 2. Prove that the mapping G(x, y) = (3x, 2y) is linear and visualize its effect
#    on a grid of points in R^2

# 3. For the following mappings, determine whether they are linear or not:
#    a. F(x, y) = (2x + 3y, x - y)
#    b. G(x, y) = (x + 2, 3y)
#    c. H(x, y) = (xy, x + y)

# 4. Find a concrete counterexample to show that mapping B is not linear.
#    That is, find specific vectors v and w such that G(v + w) ≠ G(v) + G(w).
```

```python
# Example 2: Kernel and Image of Linear Mappings
# This example demonstrates how to find the kernel and image of linear mappings

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

# ---- PART 1: DEFINITIONS ----

print("KERNEL AND IMAGE OF LINEAR MAPPINGS")
print("The kernel (or null space) of a linear mapping F: V → W is the set of all vectors")
print("in V that map to the zero vector in W: Ker(F) = {v ∈ V : F(v) = 0}")
print()
print("The image (or range) of a linear mapping F: V → W is the set of all vectors")
print("in W that are the result of applying F to some vector in V: Im(F) = {F(v) : v ∈ V}")
print()
```

```python
# ---- PART 2: DEFINING LINEAR MAPPINGS USING MATRICES ----

# Linear mappings between finite-dimensional spaces can be represented by matrices
# For F: R^3 → R^2 given by F(x, y, z) = (x + 2y - z, 3x - y + 2z)
# The matrix representation is:
A = np.array([
    [1, 2, -1],  # Coefficients of the first output component
    [3, -1, 2]   # Coefficients of the second output component
])

print("Linear mapping F: R^3 → R^2 given by F(x, y, z) = (x + 2y - z, 3x - y + 2z)")
print("Matrix representation A:")
print(A)
print()

# Function to apply the linear mapping to a vector
def linear_mapping(v):
    """Apply the linear mapping F(x, y, z) = (x + 2y - z, 3x - y + 2z)"""
    return A @ v  # Matrix-vector multiplication
```

```python

# ---- PART 3: FINDING THE KERNEL ----

print("FINDING THE KERNEL")
print("To find the kernel, we need to solve the homogeneous system of equations:")
print("x + 2y - z = 0")
print("3x - y + 2z = 0")
print()

# We can solve this system using row reduction
# Let's set up the augmented matrix [A|0]
augmented = np.column_stack((A, np.zeros((2, 1))))
print("Augmented matrix [A|0]:")
print(augmented)
print()

# Perform Gaussian elimination (manually for clarity)
print("Applying row operations:")
# Step 1: Use the first row to eliminate the first element in the second row
factor = augmented[1, 0] / augmented[0, 0]  # 3/1 = 3
augmented[1, :] = augmented[1, :] - factor * augmented[0, :]

print("After eliminating the first element in the second row:")
print(augmented)
print()

# From the resulting system:
# x + 2y - z = 0
# -7y + 5z = 0
# This gives: y = (5/7)z

# Therefore, we can parameterize the solution:
# z = t (free parameter)
# y = (5/7)t
# x = -2y + z = -2(5/7)t + t = -10/7t + t = -3/7t

print("From row reduction, we get:")
print("z = t (free parameter)")
print("y = (5/7)t")
print("x = -3/7t")
print()

print("Therefore, the kernel is:")
print("Ker(F) = {(-3/7t, 5/7t, t) : t ∈ R}")
print("This is a one-dimensional subspace (a line through the origin).")
print()

```

```python
# ---- PART 4: FINDING THE IMAGE ----

print("FINDING THE IMAGE")
print("To find the image, we need to determine which vectors in R^2 can be reached by F.")
print("This is equivalent to finding the column space of matrix A.")
print()

# One way to find the image is to compute the rank of A
rank_A = np.linalg.matrix_rank(A)
print(f"The rank of A is {rank_A}")
print(f"Since the rank is {rank_A}, the image is a {rank_A}-dimensional subspace of R^2.")
print()

# Since rank = 2, the image is the entire R^2 space
# Let's verify this by checking if we can reach any arbitrary vector in R^2

# Choose a random target vector in R^2
target = np.array([5, 7])
print(f"Let's check if the vector {target} is in the image of F")

# To do this, we need to solve the system A*v = target
# We can use numpy's least squares solver
solution, residuals, rank, s = np.linalg.lstsq(A, target, rcond=None)

print(f"Solution of the system A*v = {target}:")
print(f"v = {solution}")
print()

# Verify the solution
result = A @ solution
print(f"F(v) = A*v = {result}")
print(f"Is this equal to our target? {np.allclose(result, target)}")
print()

# If the residuals are close to zero, the target is in the image
if len(residuals) == 0 or np.allclose(residuals, 0):
    print("Since the residuals are zero, this target is in the image of F.")
    print("In fact, since rank(A) = 2, any vector in R^2 is in the image of F.")
    print("Therefore, Im(F) = R^2.")
else:
    print("The target is not in the image of F.")
print()
```

```python

# ---- PART 5: VISUALIZATION ----

# Let's visualize the kernel of F (a line in R^3)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Generate points on the kernel line
t = np.linspace(-3, 3, 100)
kernel_x = -3/7 * t
kernel_y = 5/7 * t
kernel_z = t

# Plot the kernel line
ax.plot(kernel_x, kernel_y, kernel_z, 'r-', linewidth=2, label='Kernel of F')

# Plot a few basis vectors in R^3
origin = np.zeros(3)
ax.quiver(origin[0], origin[1], origin[2], 1, 0, 0, color='blue', arrow_length_ratio=0.1, label='e1')
ax.quiver(origin[0], origin[1], origin[2], 0, 1, 0, color='green', arrow_length_ratio=0.1, label='e2')
ax.quiver(origin[0], origin[1], origin[2], 0, 0, 1, color='orange', arrow_length_ratio=0.1, label='e3')

# Plot the kernel vector for t=1
kernel_vec = np.array([-3/7, 5/7, 1])
ax.quiver(origin[0], origin[1], origin[2], kernel_vec[0], kernel_vec[1], kernel_vec[2],
          color='purple', arrow_length_ratio=0.1, label='Kernel vector for t=1')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Kernel of Linear Mapping F')

# Set limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

# Add a legend
ax.legend()

plt.show()

```

```python
# ---- PART 6: VERIFYING THE RANK-NULLITY THEOREM ----

print("VERIFYING THE RANK-NULLITY THEOREM")
print("The rank-nullity theorem states that:")
print("dim(V) = dim(Ker(F)) + dim(Im(F))")
print()

# Dimensions
dim_domain = A.shape[1]  # Number of columns = dimension of domain
dim_kernel = 1  # We found this by parameterizing the kernel
dim_image = rank_A  # Rank = dimension of image

print(f"Dimension of domain (R^3) = {dim_domain}")
print(f"Dimension of kernel = {dim_kernel}")
print(f"Dimension of image = {dim_image}")
print(f"dim(V) = dim(Ker(F)) + dim(Im(F)): {dim_domain} = {dim_kernel} + {dim_image} ✓")
print()

# ---- EXERCISES FOR STUDENTS ----

# 1. Find the kernel and image of the linear mapping G: R^3 → R^2 given by
#    G(x, y, z) = (2x - y + z, 4x - 2y + 2z)
#    Verify the rank-nullity theorem.

# 2. For the linear mapping H: R^2 → R^3 given by H(x, y) = (x, y, x+y),
#    find the kernel and image. Visualize the image of H in R^3.

# 3. Determine the kernel and image of the linear mapping F: R^3 → R^3
#    given by F(x, y, z) = (x + y, y + z, x + z).
#    What is the dimension of each?

# 4. Consider the projection mapping P: R^3 → R^3 given by P(x, y, z) = (x, y, 0).
#    Find its kernel and image, and verify the rank-nullity theorem.
```

```python
# Example 3: Matrix Representation of Linear Mappings
# This example demonstrates how to find the matrix representation of a linear mapping

import numpy as np
import matplotlib.pyplot as plt

# ---- PART 1: INTRODUCTION ----

print("MATRIX REPRESENTATION OF LINEAR MAPPINGS")
print("Every linear mapping F: V → W between finite-dimensional spaces")
print("can be represented by a matrix with respect to chosen bases.")
print()
```

```python

# ---- PART 2: FINDING THE MATRIX OF A LINEAR MAPPING ----

print("EXAMPLE 1: Finding the matrix for a simple linear mapping")
print("Consider the linear mapping F: R^2 → R^2 given by:")
print("F(x, y) = (3x - 2y, x + y)")
print()

# The standard basis for R^2 is {e1 = (1,0), e2 = (0,1)}
e1 = np.array([1, 0])
e2 = np.array([0, 1])

print("Using the standard basis {e1 = (1,0), e2 = (0,1)} for both domain and codomain:")
print()

# To find the matrix representation, we need to find the images of the basis vectors
# F(e1) and F(e2), and make these the columns of our matrix

# Define the linear mapping
def F(v):
    x, y = v
    return np.array([3*x - 2*y, x + y])

# Find the images of the basis vectors
F_e1 = F(e1)
F_e2 = F(e2)

print(f"F(e1) = F({e1}) = {F_e1}")
print(f"F(e2) = F({e2}) = {F_e2}")
print()

# Construct the matrix
A = np.column_stack((F_e1, F_e2))
print("The matrix representation A is:")
print(A)
print()

# Verify the matrix works correctly
test_vector = np.array([4, 5])
direct_result = F(test_vector)
matrix_result = A @ test_vector  # Matrix-vector multiplication

print(f"Let's verify with a test vector v = {test_vector}")
print(f"Using the function: F(v) = {direct_result}")
print(f"Using the matrix: A·v = {matrix_result}")
print(f"Do they match? {np.array_equal(direct_result, matrix_result)}")
print()
```

```python

# ---- PART 3: CHANGING BASES ----

print("EXAMPLE 2: Matrix representation with respect to different bases")
print("Let's use a different basis for R^2: {u1 = (1,1), u2 = (1,-1)}")
print()

# Define the new basis vectors
u1 = np.array([1, 1])
u2 = np.array([1, -1])

# To find the matrix representation with respect to this basis, we need:
# 1. Express the standard basis in terms of the new basis
# 2. Find the images of the new basis vectors
# 3. Express these images in terms of the new basis

# Step 1: Find the change of basis matrix P from standard to new basis
# The columns of P are the coordinates of the new basis vectors in the standard basis
P = np.column_stack((u1, u2))
print("Change of basis matrix P from standard to new basis:")
print(P)
print()

# Find the inverse of P
P_inv = np.linalg.inv(P)
print("Inverse of P (change of basis from new to standard):")
print(P_inv)
print()

# Step 2: Find the images of the new basis vectors
F_u1 = F(u1)
F_u2 = F(u2)

print(f"F(u1) = F({u1}) = {F_u1}")
print(f"F(u2) = F({u2}) = {F_u2}")
print()

# Step 3: Express these images in terms of the new basis
# We need to solve: F_u1 = c1*u1 + c2*u2 and F_u2 = d1*u1 + d2*u2
# This is equivalent to: [u1 u2] [c1; c2] = F_u1 and [u1 u2] [d1; d2] = F_u2
# So [c1; c2] = P_inv * F_u1 and [d1; d2] = P_inv * F_u2

coords_F_u1 = P_inv @ F_u1
coords_F_u2 = P_inv @ F_u2

print("Coordinates of F(u1) in the new basis:")
print(coords_F_u1)
print("Coordinates of F(u2) in the new basis:")
print(coords_F_u2)
print()

# Construct the matrix B with respect to the new basis
B = np.column_stack((coords_F_u1, coords_F_u2))
print("The matrix representation B with respect to the new basis is:")
print(B)
print()

# Verify the matrix works correctly in the new basis
# For a vector v = a*u1 + b*u2 with coordinates [a, b] in the new basis,
# the image should have coordinates B · [a, b] in the new basis

# Choose a vector in terms of the new basis
coords_v = np.array([2, 3])  # v = 2*u1 + 3*u2
print(f"Let's verify with a vector v with coordinates {coords_v} in the new basis")

# Express v in the standard basis
v_standard = P @ coords_v
print(f"In the standard basis, v = {v_standard}")

# Find the image using the original function
F_v = F(v_standard)
print(f"Using the function: F(v) = {F_v}")

# Find the image using the matrix B and convert back to standard basis
F_v_coords = B @ coords_v
F_v_standard = P @ F_v_coords
print(f"Using matrix B and converting back: {F_v_standard}")

print(f"Do they match? {np.allclose(F_v, F_v_standard)}")
print()
```

```python

# ---- PART 4: RELATIONSHIP BETWEEN MATRICES A AND B ----

print("RELATIONSHIP BETWEEN MATRICES A AND B")
print("If A is the matrix in the standard basis and B is the matrix in the new basis,")
print("the relationship is: B = P^-1 · A · P")
print()

# Calculate B using the relationship
B_calc = P_inv @ A @ P
print("B calculated using P^-1 · A · P:")
print(B_calc)
print("Original B:")
print(B)
print(f"Do they match? {np.allclose(B, B_calc)}")
print()
```

```python

# ---- PART 6: LINEAR OPERATORS (MAPPINGS FROM V TO V) ----

print("LINEAR OPERATORS")
print("A linear operator is a linear mapping from a vector space to itself.")
print("Let's consider the rotation operator in R^2.")
print()

# Define a rotation operator that rotates vectors by θ
def rotation(theta):
    """Returns a function that rotates vectors in R^2 by theta radians."""
    def rotate(v):
        c, s = np.cos(theta), np.sin(theta)
        rotation_matrix = np.array([[c, -s], [s, c]])
        return rotation_matrix @ v
    return rotate

# Create a 45-degree rotation
theta = np.pi/4  # 45 degrees
R = rotation(theta)

print(f"Rotation by {theta} radians (45 degrees):")
print(f"R(e1) = {R(e1)}")
print(f"R(e2) = {R(e2)}")
print()

# Find the matrix representation
rotation_matrix = np.column_stack((R(e1), R(e2)))
print("Matrix representation of the rotation:")
print(rotation_matrix)
print()

# Visualize the rotation
plt.figure(figsize=(8, 8))
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot standard basis vectors
plt.arrow(0, 0, e1[0], e1[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='e1')
plt.arrow(0, 0, e2[0], e2[1], head_width=0.1, head_length=0.1, fc='green', ec='green', label='e2')

# Plot their images under rotation
plt.arrow(0, 0, R(e1)[0], R(e1)[1], head_width=0.1, head_length=0.1, fc='red', ec='red', label='R(e1)')
plt.arrow(0, 0, R(e2)[0], R(e2)[1], head_width=0.1, head_length=0.1, fc='purple', ec='purple', label='R(e2)')

# Plot a grid of points and their rotated images
x = np.linspace(-1, 1, 5)
y = np.linspace(-1, 1, 5)
X, Y = np.meshgrid(x, y)

for i in range(len(X)):
    for j in range(len(X[0])):
        point = np.array([X[i,j], Y[i,j]])
        if not (point[0] == 0 and point[1] == 0):  # Skip the origin
            rotated = R(point)
            plt.plot([point[0], rotated[0]], [point[1], rotated[1]], 'k--', alpha=0.3)
            plt.scatter(point[0], point[1], c='blue', s=20)
            plt.scatter(rotated[0], rotated[1], c='red', s=20)

plt.grid(True)
plt.axis('equal')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.title('Rotation Operator (45 degrees)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()

# ---- EXERCISES FOR STUDENTS ----

# 1. Find the matrix representation of the linear mapping G: R^3 → R^2 given by
#    G(x, y, z) = (x + y + z, 2x - y + z) with respect to the standard bases.
#    Verify your matrix works correctly on a test vector.

# 2. Consider the linear operator S: R^2 → R^2 that reflects points across the line y = x.
#    Find its matrix representation with respect to the standard basis.
#    Visualize this reflection operator similar to the rotation example.

# 3. Find the matrix representation of the derivative operator D: P₂ → P₁, where
#    P₂ is the space of polynomials of degree at most 2 and P₁ is the space of
#    polynomials of degree at most 1.
#    Use the standard bases {1, t, t²} for P₂ and {1, t} for P₁.

# 4. Let F: R^2 → R^2 be the linear operator defined by F(x, y) = (2x - y, x + y).
#    Find the matrix of F with respect to the basis {(1,1), (1,-1)}.
```

```python
# Example 4: Composition and Operations on Linear Mappings
# This example demonstrates composition and other operations on linear mappings

import numpy as np
import matplotlib.pyplot as plt

# ---- PART 1: INTRODUCTION ----

print("COMPOSITION AND OPERATIONS ON LINEAR MAPPINGS")
print("Linear mappings can be combined through operations like addition,")
print("scalar multiplication, and composition.")
print()

```

```python
# ---- PART 2: DEFINING LINEAR MAPPINGS ----

# Let's define two linear mappings F, G: R^2 → R^2

# Define F(x, y) = (x + y, x - y)
def F(v):
    x, y = v
    return np.array([x + y, x - y])

# Define G(x, y) = (2x, 3y)
def G(v):
    x, y = v
    return np.array([2*x, 3*y])

# Matrix representations
A_F = np.array([[1, 1], [1, -1]])  # Matrix for F
A_G = np.array([[2, 0], [0, 3]])   # Matrix for G

print("Linear mapping F: R^2 → R^2 defined by F(x, y) = (x + y, x - y)")
print("Matrix representation of F:")
print(A_F)
print()

print("Linear mapping G: R^2 → R^2 defined by G(x, y) = (2x, 3y)")
print("Matrix representation of G:")
print(A_G)
print()
```

```python

# ---- PART 3: ADDITION OF LINEAR MAPPINGS ----

print("ADDITION OF LINEAR MAPPINGS")
print("For linear mappings F, G: V → W, the sum F + G is defined by:")
print("(F + G)(v) = F(v) + G(v) for all v in V")
print()

# Define the sum F + G
def F_plus_G(v):
    return F(v) + G(v)

# Matrix representation of F + G
A_F_plus_G = A_F + A_G

print("Matrix representation of F + G:")
print(A_F_plus_G)
print()

# Verify with a test vector
test_vector = np.array([3, 4])
sum_result_direct = F_plus_G(test_vector)
sum_result_matrix = A_F_plus_G @ test_vector

print(f"Test vector: {test_vector}")
print(f"(F + G)({test_vector}) calculated directly: {sum_result_direct}")
print(f"(F + G)({test_vector}) using matrix: {sum_result_matrix}")
print(f"Do they match? {np.array_equal(sum_result_direct, sum_result_matrix)}")
print()
```

```python

# ---- PART 4: SCALAR MULTIPLICATION ----

print("SCALAR MULTIPLICATION OF LINEAR MAPPINGS")
print("For a linear mapping F: V → W and a scalar c, the mapping cF is defined by:")
print("(cF)(v) = c·F(v) for all v in V")
print()

# Define the scalar multiple 2F
c = 2
def c_times_F(v):
    return c * F(v)

# Matrix representation of cF
A_c_times_F = c * A_F

print(f"Matrix representation of {c}F:")
print(A_c_times_F)
print()

# Verify with a test vector
c_times_result_direct = c_times_F(test_vector)
c_times_result_matrix = A_c_times_F @ test_vector

print(f"Test vector: {test_vector}")
print(f"({c}F)({test_vector}) calculated directly: {c_times_result_direct}")
print(f"({c}F)({test_vector}) using matrix: {c_times_result_matrix}")
print(f"Do they match? {np.array_equal(c_times_result_direct, c_times_result_matrix)}")
print()
```

```python

# ---- PART 5: COMPOSITION OF LINEAR MAPPINGS ----

print("COMPOSITION OF LINEAR MAPPINGS")
print("For linear mappings F: U → V and G: V → W, the composition G∘F: U → W is defined by:")
print("(G∘F)(u) = G(F(u)) for all u in U")
print()

# Define the composition G∘F
def G_compose_F(v):
    return G(F(v))

# Define the composition F∘G
def F_compose_G(v):
    return F(G(v))

# Matrix representation of G∘F and F∘G
A_G_compose_F = A_G @ A_F  # Matrix multiplication for composition
A_F_compose_G = A_F @ A_G

print("Matrix representation of G∘F:")
print(A_G_compose_F)
print()

print("Matrix representation of F∘G:")
print(A_F_compose_G)
print()

# Verify G∘F with a test vector
G_compose_F_result_direct = G_compose_F(test_vector)
G_compose_F_result_matrix = A_G_compose_F @ test_vector

print(f"Test vector: {test_vector}")
print(f"(G∘F)({test_vector}) calculated directly: {G_compose_F_result_direct}")
print(f"(G∘F)({test_vector}) using matrix: {G_compose_F_result_matrix}")
print(f"Do they match? {np.array_equal(G_compose_F_result_direct, G_compose_F_result_matrix)}")
print()

# Verify F∘G with the same test vector
F_compose_G_result_direct = F_compose_G(test_vector)
F_compose_G_result_matrix = A_F_compose_G @ test_vector

print(f"(F∘G)({test_vector}) calculated directly: {F_compose_G_result_direct}")
print(f"(F∘G)({test_vector}) using matrix: {F_compose_G_result_matrix}")
print(f"Do they match? {np.array_equal(F_compose_G_result_direct, F_compose_G_result_matrix)}")
print()

# Check commutativity
print(f"Is composition commutative? {np.array_equal(G_compose_F_result_direct, F_compose_G_result_direct)}")
print("In general, composition of linear mappings is NOT commutative: G∘F ≠ F∘G")
print()
```

```python

# ---- PART 6: INVERTIBLE LINEAR MAPPINGS ----

print("INVERTIBLE LINEAR MAPPINGS")
print("A linear mapping F: V → W is invertible if there exists a linear mapping")
print("F^(-1): W → V such that F^(-1)∘F = I_V and F∘F^(-1) = I_W,")
print("where I_V and I_W are identity mappings on V and W respectively.")
print()

# Check if F is invertible by computing the determinant of its matrix
det_F = np.linalg.det(A_F)
print(f"Determinant of matrix A_F: {det_F}")

if det_F != 0:
    print("Since the determinant is non-zero, F is invertible.")

    # Find the inverse matrix
    A_F_inv = np.linalg.inv(A_F)
    print("Matrix representation of F^(-1):")
    print(A_F_inv)

    # Define the inverse mapping
    def F_inv(v):
        return A_F_inv @ v

    # Verify that F^(-1)∘F = I
    identity = A_F_inv @ A_F
    print("F^(-1)∘F represented as a matrix (should be the identity):")
    print(identity)

    # Verify with a test vector
    F_inv_compose_F_result = F_inv(F(test_vector))
    print(f"Test vector: {test_vector}")
    print(f"(F^(-1)∘F)({test_vector}) = {F_inv_compose_F_result}")
    print(f"Is this equal to the original vector? {np.allclose(F_inv_compose_F_result, test_vector)}")
else:
    print("Since the determinant is zero, F is not invertible.")
print()
```

```python
# Example 5: Applications of Linear Mappings in Machine Learning
# This example demonstrates how linear mappings are used in machine learning

import numpy as np
import matplotlib.pyplot as plt

# ---- PART 1: INTRODUCTION ----

print("APPLICATIONS OF LINEAR MAPPINGS IN MACHINE LEARNING")
print("Linear mappings are fundamental to many machine learning algorithms.")
print("In this example, we'll explore three important applications:")
print("1. Linear Transformations in Data Preprocessing")
print("2. Linear Regression")
print("3. Principal Component Analysis (PCA)")
print()
```

```python

# ---- PART 2: LINEAR TRANSFORMATIONS IN DATA PREPROCESSING ----

print("LINEAR TRANSFORMATIONS IN DATA PREPROCESSING")
print("Data preprocessing often involves linear transformations like:")
print("- Scaling (standardization, normalization)")
print("- Whitening")
print("- Rotating")
print()

# Let's create some sample 2D data
np.random.seed(42)  # For reproducibility
mean = [2, 3]
cov = [[2, 0.5], [0.5, 1]]  # Covariance matrix
data = np.random.multivariate_normal(mean, cov, 100)

print("Sample data shape:", data.shape)
print("Sample data (first 5 points):")
print(data[:5])
print()

# Define standardization as a linear mapping
def standardize(X):
    """
    Standardizes data by subtracting the mean and dividing by the standard deviation.
    """
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    return (X - mean) / std

# Apply standardization
standardized_data = standardize(data)

print("Standardized data (first 5 points):")
print(standardized_data[:5])
print()

print("Mean of standardized data:", np.mean(standardized_data, axis=0))
print("Standard deviation of standardized data:", np.std(standardized_data, axis=0))
print()

# Let's visualize the original and standardized data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(standardized_data[:, 0], standardized_data[:, 1], alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.title('Standardized Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()

# Mathematically, standardization can be written as a linear transformation
# For each feature i: X_i' = (X_i - μ_i) / σ_i
# This isn't strictly a linear mapping (due to the mean subtraction),
# but it's an affine transformation, which is closely related
```

```python

# ---- PART 3: LINEAR REGRESSION ----

print("LINEAR REGRESSION")
print("Linear regression models a relationship between inputs X and outputs y")
print("using a linear mapping: y = X·w + b")
print("where w are weights and b is the bias term.")
print()

# Let's create some synthetic data for linear regression
np.random.seed(42)
# Input feature
X = np.random.uniform(-5, 5, 50)
# True relationship: y = 2*X + 1 + noise
y = 2 * X + 1 + np.random.normal(0, 1, 50)

# Reshape X for the model
X_design = X.reshape(-1, 1)

# Add a column of ones for the bias term
X_design_bias = np.column_stack((np.ones(X.shape), X_design))

print("Design matrix shape (with bias column):", X_design_bias.shape)
print("First few rows of the design matrix:")
print(X_design_bias[:5])
print()

# Solve the normal equations to find the weights
# (X^T X)w = X^T y
w = np.linalg.inv(X_design_bias.T @ X_design_bias) @ X_design_bias.T @ y

print("Estimated weights (bias, coefficient):", w)
print("Compared to the true values (1, 2)")
print()

# Use the model to make predictions
y_pred = X_design_bias @ w

# Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.7, label='Data points')
plt.plot(X, y_pred, 'r-', label='Fitted line')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# The linear regression model is a linear mapping F: R^n → R defined by:
# F(x) = w^T x + b
# Where x is the input feature vector, w is the weight vector, and b is the bias
```

```python

# ---- PART 4: PRINCIPAL COMPONENT ANALYSIS (PCA) ----

print("PRINCIPAL COMPONENT ANALYSIS (PCA)")
print("PCA finds a linear transformation that projects data onto")
print("a lower-dimensional subspace with maximum variance.")
print()

# Use the same data we created earlier
print("Original data shape:", data.shape)
print("Original data mean:", np.mean(data, axis=0))
print("Original data covariance matrix:")
print(np.cov(data, rowvar=False))
print()

# Step 1: Center the data (subtract the mean)
centered_data = data - np.mean(data, axis=0)

# Step 2: Compute the covariance matrix
cov_matrix = np.cov(centered_data, rowvar=False)
print("Covariance matrix of centered data:")
print(cov_matrix)
print()

# Step 3: Find the eigenvectors and eigenvalues of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
# Sort eigenvectors by eigenvalues in descending order
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
print()

# Step 4: Project the data onto the principal components
# For dimensionality reduction, we would select a subset of eigenvectors
# Here, we'll keep both dimensions to visualize the transformation
transformed_data = centered_data @ eigenvectors

print("Transformed data (first 5 points):")
print(transformed_data[:5])
print()

# Verify properties of the transformed data
print("Transformed data mean:", np.mean(transformed_data, axis=0))
print("Transformed data covariance matrix:")
print(np.cov(transformed_data, rowvar=False))
print("Note that the covariance matrix is now diagonal with eigenvalues on the diagonal.")
print()

# Visualize the original and transformed data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(centered_data[:, 0], centered_data[:, 1], alpha=0.7)
# Plot the eigenvectors scaled by their eigenvalues
for i in range(2):
    plt.arrow(0, 0, eigenvectors[0, i] * np.sqrt(eigenvalues[i]),
              eigenvectors[1, i] * np.sqrt(eigenvalues[i]),
              head_width=0.1, head_length=0.1, fc=f'C{i}', ec=f'C{i}',
              label=f'Eigenvector {i+1}')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.title('Centered Data with Principal Components')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], alpha=0.7)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.title('Data After PCA Transformation')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.tight_layout()
plt.show()

# Dimensionality reduction would involve keeping only the first k components
# Let's keep only the first principal component (reduce to 1D)
k = 1
reduced_data = transformed_data[:, :k]
print(f"Reduced data shape (keeping {k} components):", reduced_data.shape)
print("Reduced data (first 5 points):")
print(reduced_data[:5])
print()

# We can approximately reconstruct the original data from the reduced data
# by multiplying by the transpose of the eigenvector matrix (only k columns)
reconstructed_data = reduced_data @ eigenvectors[:, :k].T
print("Reconstructed data shape:", reconstructed_data.shape)
print("Reconstructed data (first 5 points):")
print(reconstructed_data[:5])
print()

# Visualize the reconstruction
plt.figure(figsize=(10, 6))
plt.scatter(centered_data[:, 0], centered_data[:, 1], alpha=0.7, label='Original data')
plt.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], alpha=0.7, label='Reconstructed data')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.grid(True)
plt.title('Data Reconstruction After PCA Dimensionality Reduction')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# The PCA transformation is a linear mapping defined by the eigenvectors:
# F(x) = P^T (x - μ)
# where P is the matrix of eigenvectors and μ is the mean of the data
```

```python

# ---- PART 5: LINEAR MAPPINGS IN NEURAL NETWORKS ----

print("LINEAR MAPPINGS IN NEURAL NETWORKS")
print("Each layer in a neural network applies a linear transformation")
print("followed by a non-linear activation function.")
print("The linear part can be represented as: h = W·x + b")
print()

# Let's implement a simple neural network with one hidden layer
def sigmoid(x):
    """Sigmoid activation function"""
    return 1 / (1 + np.exp(-x))

# Define the weights and biases
np.random.seed(42)
input_size = 2
hidden_size = 3
output_size = 1

# Initialize weights and biases
W1 = np.random.normal(0, 1, (input_size, hidden_size))
b1 = np.random.normal(0, 1, hidden_size)
W2 = np.random.normal(0, 1, (hidden_size, output_size))
b2 = np.random.normal(0, 1, output_size)

print("Weight matrix for the hidden layer (W1):")
print(W1)
print("Bias vector for the hidden layer (b1):", b1)
print()

# Define the forward pass
def forward(x):
    """Forward pass through the neural network"""
    # Linear transformation for hidden layer
    z1 = x @ W1 + b1
    # Apply activation function
    a1 = sigmoid(z1)
    # Linear transformation for output layer
    z2 = a1 @ W2 + b2
    # Apply final activation
    a2 = sigmoid(z2)
    return a2

# Generate a grid of points to visualize the neural network's decision boundary
x_min, x_max = -3, 3
y_min, y_max = -3, 3
h = 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Make predictions for all grid points
Z = np.array([forward(point) for point in grid_points])
Z = Z.reshape(xx.shape)

# Visualize the decision boundary
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdBu)
plt.colorbar(label='Network Output')
plt.title('Neural Network Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

# ---- EXERCISES FOR STUDENTS ----

# 1. Implement feature normalization (scaling to [0,1] range) as a linear transformation.
#    Apply it to the data and visualize the results.

# 2. For the linear regression problem, modify the code to perform polynomial regression
#    by adding non-linear features (e.g., X^2, X^3). Show that even though the features
#    are non-linear, the regression model itself is still a linear mapping.

# 3. Implement PCA dimensionality reduction on a dataset of your choice.
#    Experiment with keeping different numbers of principal components and
#    visualize the reconstruction error.

# 4. Create a simple linear mapping that represents a layer in a neural network.
#    Visualize how this mapping transforms a grid of points in 2D space.
```
