```python
# Example 1: Basic Vector Operations using NumPy
# This example demonstrates fundamental vector operations covered in class

import numpy as np
import matplotlib.pyplot as plt

# 1. Creating vectors
# We can create vectors as NumPy arrays
print("Creating vectors:")
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
print("Vector x =", x)
print("Vector y =", y)
print()
```

```python

# 2. Vector addition
# Adding vectors is done component-wise
print("Vector Addition:")
z = x + y
print("x + y =", z)
print()
```

```python

# 3. Scalar multiplication
# Multiplying a vector by a scalar (number)
print("Scalar Multiplication:")
scalar = 2
scaled_x = scalar * x
print(f"{scalar} * x =", scaled_x)
print()
```

```python

# 4. Dot product
# The dot product gives a scalar value
print("Dot Product:")
dot_product = np.dot(x, y)
# We can also use the @ operator: dot_product = x @ y
print("x · y =", dot_product)
print("Which equals:", x[0]*y[0] + x[1]*y[1] + x[2]*y[2])
print()
```

```python

# 5. Vector magnitude (length)
# The magnitude or length of a vector
print("Vector Magnitude:")
magnitude_x = np.linalg.norm(x)
print("|x| =", magnitude_x)
print("Which equals the square root of:", x[0]**2 + x[1]**2 + x[2]**2)
print()
```

```python

# 6. Vector normalization
# Creating a unit vector (vector with length 1) in the same direction
print("Vector Normalization:")
unit_x = x / np.linalg.norm(x)
print("Unit vector of x =", unit_x)
print("Length of unit vector:", np.linalg.norm(unit_x))
print()
```

```python
# Example 2: Solving Systems of Linear Equations using NumPy
# This example shows how to solve linear systems using NumPy's built-in functions

import numpy as np
import matplotlib.pyplot as plt

# 1. Solving a 2x2 system of linear equations
# Consider the system:
# 2x - 5y = 11
# 3x + 4y = 5

# Step 1: Define the coefficient matrix A and the constant vector b
print("Example 1: 2x2 System")
A = np.array([[2, -5],  # Coefficients of the first equation
              [3, 4]])  # Coefficients of the second equation
b = np.array([11, 5])   # Constants on the right side

# Step 2: Solve the system using numpy.linalg.solve
solution = np.linalg.solve(A, b)

# Step 3: Display the solution
print("The system:")
print("2x - 5y = 11")
print("3x + 4y = 5")
print("has the solution:")
print(f"x = {solution[0]}")
print(f"y = {solution[1]}")

# Step 4: Verify the solution by substituting back
print("\nVerifying the solution:")
print(f"2({solution[0]}) - 5({solution[1]}) = {2*solution[0] - 5*solution[1]}")
print(f"3({solution[0]}) + 4({solution[1]}) = {3*solution[0] + 4*solution[1]}")
print()
```

```python
# Example 3: Implementing Gaussian Elimination
# This example shows how to implement Gaussian elimination step by step

import numpy as np

# Define a function to perform Gaussian elimination
def gaussian_elimination(A, b):
    """
    Solves a system of linear equations using Gaussian elimination.

    Parameters:
    A -- coefficient matrix
    b -- right-hand side vector

    Returns:
    x -- solution vector
    """
    # Step 1: Create the augmented matrix [A|b]
    # We concatenate A and b horizontally
    augmented = np.column_stack((A, b))

    # Get dimensions
    n = A.shape[0]  # Number of rows

    # Step 2: Forward elimination
    for i in range(n):
        # Find pivot row (row with largest absolute value in column i)
        pivot_row = i
        for j in range(i + 1, n):
            if abs(augmented[j, i]) > abs(augmented[pivot_row, i]):
                pivot_row = j

        # Swap rows if necessary
        if pivot_row != i:
            augmented[[i, pivot_row]] = augmented[[pivot_row, i]]

        # Skip if the pivot is zero (indicates possible redundant equation)
        if abs(augmented[i, i]) < 1e-10:
            continue

        # Eliminate below the pivot
        for j in range(i + 1, n):
            factor = augmented[j, i] / augmented[i, i]
            augmented[j, i:] = augmented[j, i:] - factor * augmented[i, i:]

    # Step 3: Back substitution
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):  # Start from the last row
        # Skip if the diagonal element is zero
        if abs(augmented[i, i]) < 1e-10:
            x[i] = 0  # Assume zero for free variables
            continue

        # Calculate the value for x[i]
        sum_val = 0
        for j in range(i + 1, n):
            sum_val += augmented[i, j] * x[j]

        x[i] = (augmented[i, n] - sum_val) / augmented[i, i]

    return x

# Let's test with a simple system:
# 2x + y - z = 8
# -3x - y + 2z = -11
# -2x + y + 2z = -3

# Define the coefficient matrix A and the right side vector b
A = np.array([[2, 1, -1],
              [-3, -1, 2],
              [-2, 1, 2]], dtype=float)

b = np.array([8, -11, -3], dtype=float)

# Print the system
print("Solving the system:")
print("2x + y - z = 8")
print("-3x - y + 2z = -11")
print("-2x + y + 2z = -3")
print()

# Solve using our Gaussian elimination function
print("Step-by-step Gaussian elimination solution:")
print()

# We'll manually walk through each step to make it clear for students
augmented = np.column_stack((A, b))
print("Initial augmented matrix [A|b]:")
print(augmented)
print()

# Forward elimination - first pivot
pivot_row = 0
for j in range(1, 3):
    factor = augmented[j, 0] / augmented[0, 0]
    print(f"Eliminating x from row {j+1} using row 1:")
    print(f"Multiply row 1 by {factor:.2f} and subtract from row {j+1}")
    augmented[j] = augmented[j] - factor * augmented[0]
    print(augmented)
    print()

# Forward elimination - second pivot
pivot_row = 1
for j in range(2, 3):
    factor = augmented[j, 1] / augmented[1, 1]
    print(f"Eliminating y from row {j+1} using row 2:")
    print(f"Multiply row 2 by {factor:.2f} and subtract from row {j+1}")
    augmented[j] = augmented[j] - factor * augmented[1]
    print(augmented)
    print()

# Back substitution
x = np.zeros(3)

# Solve for z
print("Back substitution:")
print("Solving for z from the last row:")
x[2] = augmented[2, 3] / augmented[2, 2]
print(f"z = {augmented[2, 3]:.2f} / {augmented[2, 2]:.2f} = {x[2]:.2f}")
print()

# Solve for y
print("Solving for y from the second row:")
x[1] = (augmented[1, 3] - augmented[1, 2] * x[2]) / augmented[1, 1]
print(f"y = ({augmented[1, 3]:.2f} - {augmented[1, 2]:.2f} * {x[2]:.2f}) / {augmented[1, 1]:.2f} = {x[1]:.2f}")
print()

# Solve for x
print("Solving for x from the first row:")
x[0] = (augmented[0, 3] - augmented[0, 1] * x[1] - augmented[0, 2] * x[2]) / augmented[0, 0]
print(f"x = ({augmented[0, 3]:.2f} - {augmented[0, 1]:.2f} * {x[1]:.2f} - {augmented[0, 2]:.2f} * {x[2]:.2f}) / {augmented[0, 0]:.2f} = {x[0]:.2f}")
print()

# Print the solution
print("Solution found:")
print(f"x = {x[0]:.2f}")
print(f"y = {x[1]:.2f}")
print(f"z = {x[2]:.2f}")
print()

# Verify with NumPy's solver
numpy_solution = np.linalg.solve(A, b)
print("NumPy's solution for verification:")
print(f"x = {numpy_solution[0]:.2f}")
print(f"y = {numpy_solution[1]:.2f}")
print(f"z = {numpy_solution[2]:.2f}")
print()

# Verify the solution by substituting back
print("Verifying by substituting back into the original equations:")
print(f"2({x[0]:.2f}) + {x[1]:.2f} - {x[2]:.2f} = {2*x[0] + x[1] - x[2]:.2f} (should be 8)")
print(f"-3({x[0]:.2f}) - {x[1]:.2f} + 2({x[2]:.2f}) = {-3*x[0] - x[1] + 2*x[2]:.2f} (should be -11)")
print(f"-2({x[0]:.2f}) + {x[1]:.2f} + 2({x[2]:.2f}) = {-2*x[0] + x[1] + 2*x[2]:.2f} (should be -3)")

# EXERCISES FOR STUDENTS:
# 1. Modify the gaussian_elimination function to handle the case of no solutions
#    (inconsistent systems) and output an appropriate message.
#
# 2. Solve the following system using Gaussian elimination by hand, then verify with your code:
#    x + 2y + z = 2
#    2x + y - z = 1
#    x - y + z = 4
#
# 3. Create a function to transform a matrix into row echelon form without solving
#    the system completely.
```

```python
# Example 4: LU Decomposition for Solving Linear Systems
# This example demonstrates LU decomposition and how to use it to solve linear systems

import numpy as np

def lu_decomposition(A):
    """
    Performs LU decomposition on a square matrix A.

    Parameters:
    A -- square matrix to decompose

    Returns:
    L -- lower triangular matrix with 1's on diagonal
    U -- upper triangular matrix
    """
    n = A.shape[0]  # Size of the matrix

    # Initialize L and U as identity and zeros
    L = np.identity(n)
    U = np.zeros((n, n))

    # Perform LU decomposition
    for i in range(n):
        # Upper triangular part (U)
        for j in range(i, n):
            sum_val = 0
            for k in range(i):
                sum_val += L[i, k] * U[k, j]
            U[i, j] = A[i, j] - sum_val

        # Lower triangular part (L)
        for j in range(i+1, n):
            sum_val = 0
            for k in range(i):
                sum_val += L[j, k] * U[k, i]
            if abs(U[i, i]) < 1e-10:  # Check for division by zero
                raise ValueError("Matrix is singular or nearly singular")
            L[j, i] = (A[j, i] - sum_val) / U[i, i]

    return L, U

def solve_with_lu(L, U, b):
    """
    Solves a system of linear equations using LU decomposition.
    First solves Ly = b for y, then Ux = y for x.

    Parameters:
    L -- lower triangular matrix
    U -- upper triangular matrix
    b -- right-hand side vector

    Returns:
    x -- solution vector
    """
    n = L.shape[0]

    # Step 1: Solve Ly = b (forward substitution)
    y = np.zeros(n)
    for i in range(n):
        sum_val = 0
        for j in range(i):
            sum_val += L[i, j] * y[j]
        y[i] = b[i] - sum_val

    # Step 2: Solve Ux = y (backward substitution)
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        sum_val = 0
        for j in range(i+1, n):
            sum_val += U[i, j] * x[j]
        x[i] = (y[i] - sum_val) / U[i, i]

    return x
```

```python

# Example 1: Simple 3x3 system
print("EXAMPLE 1: Solving a 3x3 system using LU decomposition")
print()

# Define the coefficient matrix and right-hand side
A = np.array([[4, 3, 0],
              [3, 4, -1],
              [0, -1, 4]], dtype=float)

b = np.array([24, 30, -24], dtype=float)

print("System to solve:")
print("4x + 3y + 0z = 24")
print("3x + 4y - z = 30")
print("0x - y + 4z = -24")
print()

# Compute LU decomposition
try:
    L, U = lu_decomposition(A)

    # Display L and U matrices
    print("L matrix (Lower triangular):")
    print(L)
    print()

    print("U matrix (Upper triangular):")
    print(U)
    print()

    # Verify that LU = A
    LU = np.dot(L, U)
    print("Verification: LU should equal A")
    print("LU =")
    print(LU)
    print()
    print("Original A =")
    print(A)
    print()

    # Solve the system using LU decomposition
    x = solve_with_lu(L, U, b)

    # Print the solution
    print("Solution using LU decomposition:")
    print(f"x = {x[0]}")
    print(f"y = {x[1]}")
    print(f"z = {x[2]}")
    print()

    # Verify the solution
    print("Verification by substituting back:")
    print(f"4({x[0]}) + 3({x[1]}) + 0({x[2]}) = {4*x[0] + 3*x[1] + 0*x[2]} (should be 24)")
    print(f"3({x[0]}) + 4({x[1]}) - 1({x[2]}) = {3*x[0] + 4*x[1] - 1*x[2]} (should be 30)")
    print(f"0({x[0]}) - 1({x[1]}) + 4({x[2]}) = {0*x[0] - 1*x[1] + 4*x[2]} (should be -24)")
    print()

    # Compare with NumPy's solution
    numpy_solution = np.linalg.solve(A, b)
    print("NumPy's solution for verification:")
    print(f"x = {numpy_solution[0]}")
    print(f"y = {numpy_solution[1]}")
    print(f"z = {numpy_solution[2]}")

except ValueError as e:
    print(f"Error: {e}")
```

```python
# Example 5: Special Cases of Linear Systems
# This example demonstrates no solution, unique solution, and infinitely many solutions cases

import numpy as np
import matplotlib.pyplot as plt

# Case 1: System with a unique solution
# 2x + y = 5
# x - y = 1

print("CASE 1: SYSTEM WITH A UNIQUE SOLUTION")
print("System:")
print("2x + y = 5")
print("x - y = 1")
print()

# Define coefficient matrix and right-hand side
A1 = np.array([[2, 1], [1, -1]])
b1 = np.array([5, 1])

# Solve the system
try:
    solution1 = np.linalg.solve(A1, b1)
    print("Solution:")
    print(f"x = {solution1[0]}")
    print(f"y = {solution1[1]}")
    print()

    # Verify
    print("Verification:")
    print(f"2({solution1[0]}) + {solution1[1]} = {2*solution1[0] + solution1[1]} (should be 5)")
    print(f"{solution1[0]} - {solution1[1]} = {solution1[0] - solution1[1]} (should be 1)")
    print()
except np.linalg.LinAlgError as e:
    print(f"Error: {e}")

# Case 2: System with no solution (inconsistent)
# x + y = 2
# x + y = 3

print("CASE 2: SYSTEM WITH NO SOLUTION (INCONSISTENT)")
print("System:")
print("x + y = 2")
print("x + y = 3")
print()

# Define coefficient matrix and right-hand side
A2 = np.array([[1, 1], [1, 1]])
b2 = np.array([2, 3])

# Try to solve the system
try:
    solution2 = np.linalg.solve(A2, b2)
    print("Solution:")
    print(f"x = {solution2[0]}")
    print(f"y = {solution2[1]}")
    print()
except np.linalg.LinAlgError as e:
    print(f"Error: {e}")
    print("This system has no solution because the equations contradict each other.")
    print("The augmented matrix would have a row [0 0 | 1] after elimination,")
    print("which represents the impossible equation 0 = 1.")
    print()

# Let's solve using row reduction to see the inconsistency
augmented2 = np.column_stack((A2, b2))
print("Augmented matrix [A|b]:")
print(augmented2)
print()

print("After row reduction (row 2 = row 2 - row 1):")
augmented2[1] = augmented2[1] - augmented2[0]
print(augmented2)
print()
print("We get [0 0 | 1], which represents 0 = 1, an impossibility.")
print()

# Case 3: System with infinitely many solutions
# 2x - y = 1
# 4x - 2y = 2

print("CASE 3: SYSTEM WITH INFINITELY MANY SOLUTIONS")
print("System:")
print("2x - y = 1")
print("4x - 2y = 2")
print()

# Define coefficient matrix and right-hand side
A3 = np.array([[2, -1], [4, -2]])
b3 = np.array([1, 2])

# Try to solve the system
try:
    solution3 = np.linalg.solve(A3, b3)
    print("Solution:")
    print(f"x = {solution3[0]}")
    print(f"y = {solution3[1]}")
    print()
except np.linalg.LinAlgError as e:
    print(f"Error: {e}")
    print("This system has infinitely many solutions because the second equation")
    print("is just a multiple of the first equation (redundant).")
    print()

# Let's solve using row reduction to see the dependency
augmented3 = np.column_stack((A3, b3))
print("Augmented matrix [A|b]:")
print(augmented3)
print()

print("After row reduction (row 2 = row 2 - 2*row 1):")
augmented3[1] = augmented3[1] - 2 * augmented3[0]
print(augmented3)
print()
print("We get [0 0 | 0], which is a trivial equation 0 = 0.")
print("This means the second equation gives no new information.")
print()

print("The solution can be written as:")
print("y = 2x - 1, where x can be any real number")
print("For example, if x = 0, then y = -1")
print("If x = 1, then y = 1")
print("If x = 2, then y = 3")
print()

# Visualizing the three cases
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Case 1: Unique solution
x1 = np.linspace(-2, 4, 100)
y1_line1 = 5 - 2*x1  # 2x + y = 5 => y = 5 - 2x
y1_line2 = x1 - 1    # x - y = 1 => y = x - 1

axes[0].plot(x1, y1_line1, 'b-', label='2x + y = 5')
axes[0].plot(x1, y1_line2, 'r-', label='x - y = 1')
axes[0].scatter(solution1[0], solution1[1], color='green', s=100, label='Solution')
axes[0].grid(True, alpha=0.3)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0].set_xlim(-2, 4)
axes[0].set_ylim(-2, 4)
axes[0].set_title('Case 1: Unique Solution\nLines Intersect at One Point')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()

# Case 2: No solution
x2 = np.linspace(-2, 4, 100)
y2_line1 = 2 - x2    # x + y = 2 => y = 2 - x
y2_line2 = 3 - x2    # x + y = 3 => y = 3 - x

axes[1].plot(x2, y2_line1, 'b-', label='x + y = 2')
axes[1].plot(x2, y2_line2, 'r-', label='x + y = 3')
axes[1].grid(True, alpha=0.3)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1].set_xlim(-2, 4)
axes[1].set_ylim(-2, 4)
axes[1].set_title('Case 2: No Solution\nParallel Lines Never Intersect')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].legend()

# Case 3: Infinitely many solutions
x3 = np.linspace(-2, 4, 100)
y3_line = 2*x3 - 1   # 2x - y = 1 => y = 2x - 1

axes[2].plot(x3, y3_line, 'b-', label='2x - y = 1 and 4x - 2y = 2')
axes[2].scatter([0, 1, 2], [-1, 1, 3], color='green', s=100,
               label='Sample solutions: (0,-1), (1,1), (2,3)')
axes[2].grid(True, alpha=0.3)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[2].set_xlim(-2, 4)
axes[2].set_ylim(-2, 4)
axes[2].set_title('Case 3: Infinitely Many Solutions\nLines Coincide')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].legend()

plt.tight_layout()
plt.show()

# EXERCISES FOR STUDENTS:
# 1. Create your own examples of linear systems for each of the three cases
#    and solve them both algebraically and graphically.
#
# 2. Write a function that takes a linear system (A and b) and determines
#    whether it has a unique solution, no solution, or infinitely many solutions
#    without trying to solve it directly.
#
# 3. For a system with infinitely many solutions, write a function that
#    expresses the general solution in terms of free variables.
#
# 4. Consider the homogeneous system:
#    2x + 3y - z = 0
#    4x + 6y - 2z = 0
#    Determine all solutions and explain whether this is a case of no solution,
#    unique solution, or infinitely many solutions.
```
